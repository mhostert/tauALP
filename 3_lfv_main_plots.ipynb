```python
import numpy as np
from alp import plot_tools as pt
```

```python
fig, ax = pt.std_fig(figsize=(5,6))

name = 'anarchy'
lamb = 1
c_lepton = np.array([[1, lamb, lamb], 
                     [lamb, 1, lamb],
                     [lamb, lamb, 1]
                     ])

pt.plot_other_limits(ax, c_lepton=c_lepton, linewidth=0.1, c_NN=0, mN=0.0, annotate=True)

ax.set_yscale('log')
ax.set_xscale('log')
fa_power=1

ax.set_ylabel(rf"$f_a^{{{-fa_power}}}$ [GeV$^{{{-fa_power}}}$]")
ax.set_xlabel(r"$m_a$ [GeV]")

title = r"{\bf LFV anarchy} $\,\,\vert\,\,$"
# ax.annotate(r'\noindent \bf LFV anarchy', xy=(0.02, 0.98), xycoords='axes fraction', fontsize=11, ha='left', va='top')
title += rf"$g_{{\ell_1 \ell_2}} = {int(c_lepton[0,0])}$"
ax.set_title(title, fontsize=11, pad=7.5)
# 

ax.set_ylim(1e-11,1e-3)
ax.set_xlim(1e-2,2)
fig.savefig('plots/ALP_anarchy_otherlimits.pdf', bbox_inches='tight', dpi=300)
```

# Anarchical LFV

```python
BP_NAME = 'anarchy_jul19'
c_lepton = np.ones((3,3))
pt.main_plot_LFV(BP_NAME, c_lepton, 0, 0, ymin=1e-11, figsize=(5,6), loc='lower center', xmin=1e-2, linewidth=0.1, annotate=True, ncol=3, vlines=False)
pt.main_plot_LFV(BP_NAME, c_lepton, 0, 0, figsize=(5,6), fa_power=1, ymin=2e-10, ymax=3e-8, xmin=0.1, xmax=1.8, ncol=3, loc='lower center', xscale='linear', legend=True, name_modifier='_fa4', vlines=False, linewidth=1, grid=False)
```

## Hierarchical LFV

```python
for lam in [0.05,0.01,0.005,0.001]:
    BP_NAME = f'hierarchy_lambda_{lam}_clep_1_jul19'
    lamb = lam
    C_LEPTON = 1
    c_lepton = C_LEPTON*np.array([[1, lamb, lamb], 
                                  [lamb, 1, lamb], 
                                  [lamb, lamb, 1]])
    pt.main_plot_LFV(BP_NAME, c_lepton, 0, 0, ymin=2e-10, ymax=1e-3, loc='lower left', xmin=1e-2, linewidth=0.1, annotate=True, ncol=3, figsize=(5,6), vlines=False)
    
```

### Comparison to Bertuzzo et al
